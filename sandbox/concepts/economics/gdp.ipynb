```julia
using CSV, DataFrames, Plots
```

```julia
data = []
open("./data/gdp.csv", "r") do f
    while ! eof(f)
        line = readline(f)
        line = replace(line, "\""=>"")
        line = strip(line, [','])
        push!(data, split(line, ","))
    end
end
```

```julia
plotData = []
for i in 6:length(data)
    gdp = []
    for word in data[i][5:length(data[i])]
        num = 0
        try
            num = parse(Float64, word)
        catch
        end
        push!(gdp, num)
    end
    push!(plotData, gdp)
end
plot(plotData)
```
