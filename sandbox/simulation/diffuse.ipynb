```julia
xmax = 256
ymax = 256
xmid = Int(floor(xmax/2))
ymid = Int(floor(ymax/2))
points = [[1,.9], [.5,.5]]
m = fill(0.0, xmax, ymax)
dist(x, y, x′, y′) = sqrt((x′-x)^2 + (y′-y)^2)
for x ∈ 1:xmax
    for y ∈ 1:ymax
        val = 0
        for point ∈ points
            val -= dist(x,y,point[1]*xmax,point[2]*ymax)
        end
        m[x,y] = val
    end
end
heatmap(m)
```
