```julia
using Plots
```

```julia
tmin = 0
tmax = 4π
tvec = range(tmin, tmax, length = 100)
```

```julia
ts = 0
Δt = 0.1π
te = 2π
anim = @animate for i ∈ ts:Δt:te
    plot(1:.1:10, sin.((1:.1:10) .+ i))
end
gif(anim, "sin.gif", fps=10)
```

```julia
elements = rand(1:10, 100)

anim = @animate for (i, element) ∈ enumerate(elements)
    cur = i
    while cur != 1
        if elements[cur] < elements[cur-1]
            plot(1:length(elements), elements)
            elements[cur], elements[cur-1] = elements[cur-1], elements[cur]
        else
            break
        end
        cur -= 1
    end
end
gif(anim, "sort.gif", fps=10)
```
